1. **Menghitung Limit Menggunakan MATLAB**

Untuk menghitung limit persamaan $$f_1(x) = \lim_{x \to 0} \frac{\sin(x)}{x}$$ dan $$f_2(x) = \lim_{h \to 0} \frac{\cos(x + h) - \cos(x)}{h}$$, kita dapat menggunakan fungsi `limit` di MATLAB. Berikut adalah kode MATLAB yang sesuai:

```matlab
syms x h
f1 = sin(x)/x;
limit(f1, x, 0)

f2 = (cos(x + h) - cos(x))/h;
limit(f2, h, 0)
```

Dari kode di atas, kita dapat mengetahui bahwa $$\lim_{x \to 0} \frac{\sin(x)}{x} = 1$$ dan $$\lim_{h \to 0} \frac{\cos(x + h) - \cos(x)}{h} = -\sin(x)$$.

2. **Membuat Plot Menggunakan `Plots.jl` di Julia**

Untuk membuat plot dari fungsi tersebut, kita dapat menggunakan kode yang Anda berikan sebelumnya dengan sedikit modifikasi. Kita perlu mengganti fungsi `Z` dengan fungsi limit yang telah kita hitung sebelumnya.

````julia
using Plots

# Mendefinisikan rentang nilai x dan h
x = range(-2, stop=2, length=100)
h = range(-2, stop=2, length=100)

# Membuat grid dengan pendekatan broadcasting
X = x' .* ones(length(h))
H = ones(length(x))' .* h

# Menghitung nilai fungsi limit
Z1 = sin.(X) ./ X
Z2 = (cos.(X .+ H) .- cos.(X)) ./ H

# Menggambar grafik
p1 = surface(X, H, Z1, title="Grafik f1(x) = sin(x)/x", xlab="x", ylab="h", zlabel="f1(x)")
p2 = surface(X, H, Z2, title="Grafik f2(x) = (cos(x + h) - cos(x))/h", xlab="x", ylab="h", zlabel="f2(x)")

plot(p1, p2, layout = (2, 1))
````
