1. Dengan menggunakan matlab / Symbolab  gambarlah grafik dari persamaan berikut


````julia
using Plots

# Mendefinisikan rentang nilai x dan y
x = range(-2π, stop=2π, length=100)
y = range(-2, stop=2, length=100)

# Membuat grid dengan pendekatan broadcasting
X = x' .* ones(length(y))
Y = ones(length(x))' .* y

# Menghitung nilai fungsi
Z = 2 .* Y .* sin.(2 .* X)

# Menggambar grafik
surface(X, Y, Z, title="Grafik f(x,y) = 2y sin(2x)", xlab="x", ylab="y", zlabel="f(x,y)")
````


# Penjelasan Sederhana

Kode ini digunakan untuk menggambar grafik dari fungsi $$f(x,y) = 2y \sin(2x)$$ menggunakan dua pustaka grafik Julia, yaitu `Plots.jl` dan `CairoMakie`.

Berikut adalah penjelasan langkah-langkahnya:

1. **Mendefinisikan Rentang Nilai x dan y**: Kode ini mendefinisikan rentang nilai untuk variabel x dan y. Rentang ini ditentukan dari -2π hingga 2π dengan jumlah titik sebanyak 100.

```julia
x = range(-2π, stop=2π, length=100)
y = range(-2, stop=2, length=100)
```

2. **Membuat Grid**: Kode ini membuat grid dua dimensi dari nilai x dan y menggunakan pendekatan broadcasting. Broadcasting adalah cara untuk melakukan operasi pada array dengan ukuran yang berbeda.

```julia
X = x' .* ones(length(y))
Y = ones(length(x))' .* y
```

3. **Menghitung Nilai Fungsi**: Kode ini menghitung nilai fungsi $$f(x,y) = 2y \sin(2x)$$ untuk setiap titik pada grid.

```julia
Z = 2 .* Y .* sin.(2 .* X)
```

4. **Menggambar Grafik**: Kode ini menggambar grafik 3D dari fungsi tersebut. Dalam `Plots.jl`, fungsi `surface` digunakan, sedangkan dalam `CairoMakie`, fungsi `CairoMakie.surface` digunakan.

```julia
surface(X, Y, Z, title="Grafik f(x,y) = 2y sin(2x)", xlab="x", ylab="y", zlabel="f(x,y)")
```

```julia
CairoMakie.surface(X, Y, Z, colormap=:viridis)
```

Perbedaan utama antara kedua pustaka ini adalah cara mereka merender grafik. `Plots.jl` lebih umum dan fleksibel, sedangkan `CairoMakie` berfokus pada output grafik vektor melalui Cairo.jl. Dalam kode ini, `CairoMakie` juga menggunakan colormap `:viridis` untuk pewarnaan grafik. Colormap adalah skema warna yang digunakan untuk memetakan nilai numerik ke warna dalam visualisasi data. Colormap `:viridis` adalah salah satu pilihan colormap yang populer karena memberikan kontras yang baik dan mudah dilihat oleh orang dengan daltonisme.


````julia
using Pkg
Pkg.add("Plots")
````

````julia
using Plots

# Mendefinisikan rentang nilai x dan h
x = range(-2π, stop=2π, length=100)
h = range(-2, stop=2, length=100)

# Membuat grid dengan pendekatan broadcasting
X = x' .* ones(length(h))
H = ones(length(x))' .* h

# Menghitung nilai fungsi limit
Z1 = sin.(X) ./ X
Z2 = (cos.(X .+ H) .- cos.(X)) ./ H

# Menggambar grafik
p1 = surface(X, H, Z1, title="Grafik f1(x) = sin(x)/x", xlab="x", ylab="h", zlabel="f1(x)")
p2 = surface(X, H, Z2, title="Grafik f2(x) = (cos(x + h) - cos(x))/h", xlab="x", ylab="h", zlabel="f2(x)")

plot(p1, p2, layout = (2, 1))
````

````julia
using CairoMakie

# Mendefinisikan rentang nilai x dan h
x = range(-2π, stop=2π, length=100)
h = range(-2, stop=2, length=100)

# Membuat grid dengan pendekatan broadcasting
X = x' .* ones(length(h))
H = ones(length(x))' .* h

# Menghitung nilai fungsi limit
Z1 = sin.(X) ./ X
Z2 = (cos.(X .+ H) .- cos.(X)) ./ H

# Menggambar grafik
fig = Figure(resolution = (600, 400))
ax1 = Axis3(fig[1, 1])
surface!(ax1, X, H, Z1, colormap=:viridis)
ax1.title = "Grafik f1(x) = sin(x)/x"
ax1.xlabel = "x"
ax1.ylabel = "h"
ax1.zlabel = "f1(x)"

ax2 = Axis3(fig[2, 1])
surface!(ax2, X, H, Z2, colormap=:viridis)
ax2.title = "Grafik f2(x) = (cos(x + h) - cos(x))/h"
ax2.xlabel = "x"
ax2.ylabel = "h"
ax2.zlabel = "f2(x)"

fig
````
