```julia
using Plots

# Mendefinisikan rentang nilai x
x = range(-10, stop=10, length=1000)

# Mendefinisikan fungsi
y1 = x.^2 .- 16
y2 = x.^2 .- 4 .* x .+ 3
y3 = .- x.^2 .+ 2 .* x .+ 3

# Membuat plot
p = plot(x, y1, label="y = x^2 - 16", lw=2)
plot!(p, x, y2, label="y = x^2 - 4x + 3", lw=2)
plot!(p, x, y3, label="y = -x^2 + 2x + 3", lw=2)

# Menampilkan plot
p
```

3. Diketahui fungsi berikut f(x) =


$$
f(x) = 
\begin{cases} 
x + 3 & \text{untuk } x < 1 \\
-x + 2 & \text{untuk } x \geq 2 
\end{cases}
$$

Berikut adalah solusinya:

a. $$\lim_{{x \to -1}} f(x)$$
Karena -1 kurang dari 1, kita menggunakan bagian pertama dari fungsi, yaitu $$f(x) = x + 3$$. Jadi, kita mendapatkan:

$$\lim_{{x \to -1}} f(x) = (-1) + 3 = 2$$

b. $$\lim_{{x \to -2}} f(x)$$
Karena -2 juga kurang dari 1, kita menggunakan bagian pertama dari fungsi lagi. Jadi, kita mendapatkan:

$$\lim_{{x \to -2}} f(x) = (-2) + 3 = 1$$

c. $$\lim_{{x \to 1^-}} f(x)$$
Ini adalah limit sepihak dari kiri di x = 1. Karena kita mendekati 1 dari kiri, kita menggunakan bagian pertama dari fungsi. Jadi, kita mendapatkan:

$$\lim_{{x \to 1^-}} f(x) = (1) + 3 = 4$$



4. Hitunglah limit tak hingga berikut


a. $$\lim_{{x \to -\infty}} \frac{x^2 - 2x + 5}{2x + 5}$$
b. $$\lim_{{x \to -\infty}} \frac{-3x^2}{2x^2 + 3x}$$

Berikut adalah solusinya:

a. Untuk menyelesaikan limit ini, kita bisa membagi setiap suku di pembilang dan penyebut dengan \(x^2\) (karena \(x^2\) adalah suku dengan pangkat tertinggi), kemudian kita hitung limitnya ketika \(x\) mendekati \(-\infty\):

$$\lim_{{x \to -\infty}} \frac{x^2 - 2x + 5}{2x + 5} = \lim_{{x \to -\infty}} \frac{1 - \frac{2}{x} + \frac{5}{x^2}}{\frac{2}{x} + \frac{5}{x^2}}$$

Ketika \(x\) mendekati \(-\infty\), suku-suku yang memiliki \(x\) di penyebut akan mendekati 0. Jadi, limit ini adalah \(\frac{1}{0}\), yang tidak terdefinisi.

b. Sama seperti sebelumnya, kita membagi setiap suku di pembilang dan penyebut dengan \(x^2\), kemudian kita hitung limitnya ketika \(x\) mendekati \(-\infty\):

$$\lim_{{x \to -\infty}} \frac{-3x^2}{2x^2 + 3x} = \lim_{{x \to -\infty}} \frac{-3}{2 + \frac{3}{x}}$$

Ketika \(x\) mendekati \(-\infty\), suku \(\frac{3}{x}\) akan mendekati 0. Jadi, limit ini adalah \(-\frac{3}{2}\).
